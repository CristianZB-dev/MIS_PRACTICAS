### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
x = np.linspace(0,10,100) # establecemos la variable x que va a albergar 100 números que van del 0 al 10(rango)

y1 = np.sin(x) # calculamos el seno de cada número anteriormente creado para tener unos datos con los que el programa posteiormente presentará
y2 = np.cos(x) # calculamos el coseno de cada número de la "carpeta x" para poder representar en la gráfica.

plt.subplot( 1, 2, 1) # Definimos el primer subplot, es decir, una figura  con 1 fila y 2 columnas. Seleccionamos la columna 1 para poner nuestro gráfico.
plt.plot(x, y1) # dibujamos la curva del seno en el primer subplot que hemos seleccionado en la línea de código 
plt.title ("Seno")

# REPETIMOS LO ANTERIOR PERO DIBUJANDO EL COSENO Y ASIGNANDO LA FIGURA EN LA POSICIÓN 2 DE NUESTRO SUBPLOT

plt.subplot( 1, 2, 2) # Seleccionamos la posición 2º para nuestra figura entro de nuestro "tablero"
plt.plot(x, y2) # Dibujamos la curva del coseno en la posición 2º. X representa los datos base mientras que y2 representa 
plt.title("Coseno")

# AJUSTAMOS LOS GRÁFICOS DE MANERA AUTOMÁTICA 
plt.tight_layout()

# MOSTRAMOS FIGURAS AJUSTADAS
plt.show()

```

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
y3 = np.tan(x) # x son los datos base y el resto la operación que queremos hacer con ellos, en este caso la tangente.
```

```python
# Necesitamos crear una figura que englobe 3 subplots en la primera columna . Para ello hacemos lo siguiente:

fig, axs = plt.subplots(3,1) # Creación de 3 figuras en una columna.  Esto nos dará 3 axs que se refieren a las figuras situadas verticalmente. AXS[0] se refiere a la primera figura mientras que la 1 a la segunda y la 2 a la tercera( EL ORDEN EMPIEZA SIEMPRE EN 0)

axs[0].plot(x, y1,"r") # x es el conjunto de datos base que tenemos para sacar el Seno. Y1 son los datos pasados a senos , es decir , y1 = (x,np.sin(x))  | Se utiliza "r" para que pinte la finción en rojo(red)
axs[0].set_title("Seno") # asignamos el título para no confundirnos y darnos mayor claridad.

axs[1].plot(x, y2, "g") # Lo mismo que antes, x es el conjunto de datos, y2 no referimos al cálculo que hemos hecho antes para obtener el coseno de los números base guardados en x , es decir, y2 = (x,np.cos(x)) | "g" es para pintar la función de verde(green)
axs[1].set_title("Coseno") # asignamos el titulo a la función.

# para hacer la tangente de los datos en x hacemos lo siguiente:

axs[2].plot(x, y3, "b") # pintamos la función de la tangente en la tercera figura disponible en nuestro tablero. Color azul.
axs[2].set_title("Tangente") # título 

# AJUSTE AUTOMÁTICO DE REPRESENTACIÓN DE LAS FUNCIONES
plt.tight_layout()

# MOSTRAMOS FIGURA
plt.show()

```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperbólico y coseno hiperbólico se obtienen así:
* np.sinh, seno hiperbólico;
* np.cosh, coseno hiperbólico;


```python
# Necesitamos crear una figura de 1 columna y dos filas.
fig, (axs1,axs2) = plt.subplots(2,1) # Creación de 2 figuras en una columna.  Esto nos dará 2 axs que se refieren a las figuras situadas verticalmente. AXS[0] se refiere a la primera figura mientras que la 1 a la segunda.


# AXS 1 Nos referimos al primer plot de nuestra FIG GENERAL. 
axs1.plot(x,np.sinh(x),"r") # señalamos lo que queremos mostrar en el primer plot. x es el conjunto de datos base para nuestros cálculos. np.sinh(x) es el cálculo del seno hiperbólico.
axs1.set_title("Seno Hiperbólico") # configuramos el título del plot . Primero nombramos el plot que nos interesa modificar(AXS1) y utilizamos el método .set_title("")
axs1.set_xlabel("x") # configuramos el título del eje x mediante .set_xlabel("")
axs1.set_ylabel("sinh(x)") # configuramos el título del eje y mediante .set_ylabel("")


# AXS 2 Nos referimos al primer plot de nuestra FIG GENERAL. 
axs2.plot(x,np.cosh(x),"b")
axs2.set_title("Coseno Hiperbólico")
axs2.set_xlabel("x")
axs2.set_ylabel("cosh(x)")


plt.tight_layout()
plt.show()
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosición (tal y como ves en la figura), emplea `subplots_adjust` y los parámetros hspace y wspace.

```python
# primero identificamos qué representan los gráficos. 
        # 0,0 es el seno de x
        # 0,1 es el coseno de x
        # 1,0 es la tangente de x
        # 1,1 es el arcotangente de x


fig, axs = plt.subplots(2,2)  # Creamos el plot principal donde vamos a albergar los 4 subplots.

x = np.linspace(0,10,100) # creamos los valores base con los que vamos a trabajar. 100 valores del 0 al 10.
y1 = np.sin(x)     # SENO        |  y1 almacenará el calculo del seno de todos los números de x
y2 = np.cos(x)     # COSENO      |  y2 almacenará el calculo del coseno de todos los números de x
y3 = np.tan(x)     # TANGENTE    |  y3 almacenará el calculo de todas las tangentes de los numeros de x
y4 = np.arctan(x)  #ARCOTANGENTE |  y4 alamcenará el calcul de todos las arcotangentes de los numeros de x 


axs[0,0].plot(x,y1,"b") # pintamos seno en el primer cuadrante
axs[0,0].set_title("SENO")

axs[0,1].plot(x,y2,"b") # pintamos coseno en el segundo cuadrante
axs[0,1].set_title("COSENO")

axs[1,0].plot(x,y3,"b") # pintamos tangente en el tercer cuadrante
axs[1,0].set_title("TANGENTE")

axs[1,1].plot(x,y4,"b"); # pintamos arcotangente en el cuarto cuadrante
axs[1,1].set_title("ARCOTAGENTE"); # ponemos ; para que no nos salgan números raros al principio del plot

plt.tight_layout() # ajustamos los titulos para que no se sobrepongan

# Vamos a ajustar los subplots como nos manda el ejercicio mediante "subplot_adjust"

plt.subplots_adjust(hspace=0.5,wspace=0.5)
plt.show()




```

# EXPLICACIÓN ADICIONAL: ¿ PARA QUÉ SIRVE "plt.tight_layout()" AL FINAL DE LA CREACIÓN DE UN PLOT?

```python
import matplotlib.pyplot as plt
import numpy as np # Importamos las herramientas para la creación de nuestro plot

x = np.linspace(0, 10, 100) # creamos datos de prueba con los que vamos a hacer la presentación. En este caso mostrará 100 números en un rango del 0 al 10.
y = np.sin(x) # calculamos el seno de los número que hemos sacado en la línea anterior.

fig, axs = plt.subplots(2, 1) # cremaos el plot principal . Este va a tener 2 filas en 1 columna.

axs[0].plot(x, y) # asignamos para el primer subplot la x que son los datos base y la y que representa los números que salen del cálculo del seno de x(los datos originales)
axs[0].set_title("Gráfico del Seno con etiqueta muy larga que puede superponerse")
axs[0].set_ylabel("Valor del seno")

axs[1].plot(x, np.cos(x))
axs[1].set_title("Gráfico del Coseno con etiqueta también bastante extensa")
axs[1].set_ylabel("Valor del coseno")
axs[1].set_xlabel("Eje x con valores entre 0 y 10")

# plt.tight_layout()  # ❌ Intenta comentar y descomentar esta línea

plt.show()
```

* Si no ponemos "plt.tight_layout()" los títulos de los subplots se ven sobrepuestos a otros datos y ajustarlos de manera manual es muy tedioso. Para que se ajuste de manera automática, necesitamos usar "plt.tight_layout()".

```python
import matplotlib.pyplot as plt
import numpy as np # Importamos las herramientas para la creación de nuestro plot

x = np.linspace(0, 10, 100) # creamos datos de prueba con los que vamos a hacer la presentación. En este caso mostrará 100 números en un rango del 0 al 10.
y = np.sin(x) # calculamos el seno de los número que hemos sacado en la línea anterior.

fig, axs = plt.subplots(2, 1) # cremaos el plot principal . Este va a tener 2 filas en 1 columna.

axs[0].plot(x, y) # asignamos para el primer subplot la x que son los datos base y la y que representa los números que salen del cálculo del seno de x(los datos originales)
axs[0].set_title("Gráfico del Seno con etiqueta muy larga que puede superponerse")
axs[0].set_ylabel("Valor del seno")

axs[1].plot(x, np.cos(x))
axs[1].set_title("Gráfico del Coseno con etiqueta también bastante extensa")
axs[1].set_ylabel("Valor del coseno")
axs[1].set_xlabel("Eje x con valores entre 0 y 10")

plt.tight_layout()  # ❌

plt.show()
```
