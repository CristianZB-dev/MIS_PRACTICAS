### Ejercicio 1

Completa el gráfico con los textos necesarios para reconstruir la figura.

```python
# Lo primero definomos las variables. En este caso "x" e "y"
x = [1,2,3,4,5]
y = [2,3,5,7,11]

# Pintamos las dos variables 
plt.plot(x,y,"red", marker = "x")  # los valores de "x"en el eje horizontal y los valores de "y" en el eje vertical. Marker lo utilizamos para pintar donde se cortan las dos variables. "red" para pintar la línea.
plt.title("Ejercicio 1#")
plt.xlabel("X")
plt.ylabel("Y")

# Colocamos las coordenadas de los puntos en nuestro gráfico.

for i in range(len(x)):  # para todos los valores que tenemos en x, es decir,"x = [1,2,3,4,5]""
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})')
```

### Ejercicio 2

Completa el ejercicio (tendrás que revisar los argumentos de la función/método `text`) para reproducir la imagen de la figura:

```python
plt.plot(x,y, marker = "o") # dibujamos la gráfica simplificada para posteriormente ir introduciendo los detalles. En los puntos de corte metemos los marcadores * como "marker = "o".

plt.title("Texto Alineado") # colocamos el título.
plt.xlabel("X")
plt.ylabel("Y")

# En cuando a pintar las coordenadas de los puntos de corte, volvemos a crear el bucle "for" de antes.

for i in range (len(x)): # Para todos los valores de nuestra lista "X"
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})')  # en las coordenadas de x[i] e y[i], pintamos "f'({x[i]}, {y[i]})"
```


### Ejercicio 3

Completa el código para poder reproducir la imagen de la figura, tendrás que revisar el parámetro "connectionstyle" del rango de parámetros del argumento "arrowprops". 
Nota: Las coordenadas del texo no están dadas como puntos de la gráfica sino como "offset" del punto señalado. Esto quiere decir que si el offset es la tupla (-2,10) y las coordenadas (xy) del punto señalado por la anotación son (30,40), por ejemplo, el texto se situará en (30-2, 40 + 10) es decir en (28,50) [Recuerda, es la coordenada de la esquina inferior izquierda del cuadro de texto "invisible" que contiene al texto]

```python

plt.plot(x, y, marker='o') # Pinta la gráfica teniendo en cuenta los datos de x e y . Añadimos un marcador en cada punto de corte.
coordenadas = (x[2],y[2]) # Establecemos dónde queremos señalar la flecha. x[2] nos lleva al tercer valor la lista , en este caso el "3". Lo mismo pasa con y[2] que es igual a "5"

plt.annotate("Punto destacado", xy =coordenadas, textcoords='offset points', xytext=(-25, 40), # plt.annotate nos sirve para hacer anotaciones en nuestro plot como texto o flechas.
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2')) # "punto destacado" es el texto que queremos que aparezca. xy = coordenadas son las coordenadas donde queremos que apareza. "xytext" es lo que desplazamos el texto, en este caso -25 mueve el texto 25 puntos a la izquierda y 40 hacia arriba.
                                                                                # arrowprops = dict(arrowstyle = "->, connectionstyle = "arc3,rad=.2) es tanto el estilo de la flecha como la direción hacia donde tiene que moverse.
plt.title('Anotación Básica')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

```


### Ejercicio 4

Siguiendo el estilo de offset del ejercicio anterior, completa el código siguiente para poder reproducir la imagen de la figura:




```python
plt.plot(x, y, marker='o')

# Aquí los annotate
coordenadas_inicio = (x[0],y[0])
coordenadas_final = (x[4],y[4])
plt.annotate("Inicio", xy = coordenadas_inicio,textcoords='offset points', xytext=(100, -5),    
             arrowprops=dict(facecolor='black'))
plt.annotate("Final", xy = coordenadas_final,textcoords='offset points', xytext=(-50, -25),    
             arrowprops=dict(facecolor='red'))



plt.title('Ejercicio 4')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Ejercicio 5

Carga el dataset de accidentes de tráfico en EEUU que está en la ruta "./data/car_crashes.csv". Quedate con los cinco estados con menos accidentes en "total".

```python
df_car_crashes = pd.read_csv("./data/car_crashes.csv", sep = "|") # cargamos dataset
df_car_crashes.head() # comprobamos si se ha cargado bien
```

```python
df_car_crashes = df_car_crashes.drop(columns=["Unnamed: 0"])  # eliminamos "Unnamed: 0" porque no nos aporta nada, solo ruido a nuestros datos.
```

```python
df_car_crashes.head() 
```

```python
accientes_por_estado = df_car_crashes.groupby("abbrev")["total"].sum().sort_values().reset_index() # agrupamos por Estado el número total de accidentes. En () ponemos el valor que agrupará. [] ponemos el campo que se tratará para hacer dichas agrupaciones. "sort_values" lo usamos para ordenar de menor a mayor.
top_5_estados_menos_accidentes = accientes_por_estado.head(5)                                       # Reset_index lo usamos para que el indice no se coma ninguna columna. De no poner esto la columna "abbrev" no se identificaría.
top_5_estados_menos_accidentes
```

### Ejercicio 6

Completa el siguiente código para obtener una figura como la representada en la imagen

```python
top_5_estados_menos_accidentes = pd.DataFrame(top_5_estados_menos_accidentes)
top_5_estados_menos_accidentes.columns
```

```python
#OJO df_top_min contiene el subdataframe resultado del ejercicio anterior
estados = {
    "DC": "District of Columbia",
    "MA": "Massachusetts",
    "MN": "Minnesota",
    "__": "Washington",
    "CT": "Connecticut"
}
for abrev,estado in estados.items():
    top_5_estados_menos_accidentes["abbrev"] = top_5_estados_menos_accidentes["abbrev"].str.replace(abrev,estado) # Utilizamos el bucle "for" para sustituir las abreviaturas por el nombre completo de los estados que hemos configurado en nuestro diccionario.
colores = ["green","gray","orange","blue","pink"] # establecemos la paleta de colores que queremos utilizar en nuestro gráfico.
plt.figure(figsize = (15,5)) # configuramos nuestro lienzo , donde pintaremos nuestro gráfico de barras.
plt.bar(top_5_estados_menos_accidentes["abbrev"], height = top_5_estados_menos_accidentes["total"], color = colores, alpha = 0.5) # plt.bar para crear un plot de barras. El primer componente sirve para decir qué es lo que definirá cada una de las barras[Estados]. Height para establecer el alto de estas barras con los datos de los acciendetes(total de accidentes). Finalmente se establecen los colores y alpha el parámetro de transparencia.
plt.ylabel("Total Accidentes por Millones de Millas Recorridas") 
plt.ylim(0,30) # establecemos el límite del eje Y 
plt.axhline(y = df_car_crashes.total.max()) # pintamos una linea con plt.axhline. Dicha linea viene dada por el valor máximo de accidentes .
plt.title("Ejercicio 6"); # Titulo de nuestro gráfico de barras
```
