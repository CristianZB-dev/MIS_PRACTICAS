### Ejercicio 1

Completa el grfico con los textos necesarios para reconstruir la figura.

```python
# Lo primero definomos las variables. En este caso "x" e "y"
x = [1,2,3,4,5]
y = [2,3,5,7,11]

# Pintamos las dos variables 
plt.plot(x,y,"red", marker = "x")  # los valores de "x"en el eje horizontal y los valores de "y" en el eje vertical. Marker lo utilizamos para pintar donde se cortan las dos variables. "red" para pintar la lnea.
plt.title("Ejercicio 1#")
plt.xlabel("X")
plt.ylabel("Y")

# Colocamos las coordenadas de los puntos en nuestro grfico.

for i in range(len(x)):  # para todos los valores que tenemos en x, es decir,"x = [1,2,3,4,5]""
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})')
```

### Ejercicio 2

Completa el ejercicio (tendrs que revisar los argumentos de la funcin/mtodo `text`) para reproducir la imagen de la figura:


### Ejercicio 3

Completa el cdigo para poder reproducir la imagen de la figura, tendrs que revisar el parmetro "connectionstyle" del rango de parmetros del argumento "arrowprops". 
Nota: Las coordenadas del texo no estn dadas como puntos de la grfica sino como "offset" del punto sealado. Esto quiere decir que si el offset es la tupla (-2,10) y las coordenadas (xy) del punto sealado por la anotacin son (30,40), por ejemplo, el texto se situar en (30-2, 40 + 10) es decir en (28,50) [Recuerda, es la coordenada de la esquina inferior izquierda del cuadro de texto "invisible" que contiene al texto]

```python

plt.plot(x, y, marker='o') # Pinta la grfica teniendo en cuenta los datos de x e y . Aadimos un marcador en cada punto de corte.
coordenadas = (x[2],y[2]) # Establecemos dnde queremos sealar la flecha. x[2] nos lleva al tercer valor la lista , en este caso el "3". Lo mismo pasa con y[2] que es igual a "5"

plt.annotate("Punto destacado", xy =coordenadas, textcoords='offset points', xytext=(-25, 40), # plt.annotate nos sirve para hacer anotaciones en nuestro plot como texto o flechas.
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2')) # "punto destacado" es el texto que queremos que aparezca. xy = coordenadas son las coordenadas donde queremos que apareza. "xytext" es lo que desplazamos el texto, en este caso -25 mueve el texto 25 puntos a la izquierda y 40 hacia arriba.
                                                                                # arrowprops = dict(arrowstyle = "->, connectionstyle = "arc3,rad=.2) es tanto el estilo de la flecha como la direcin hacia donde tiene que moverse.
plt.title('Anotacin Bsica')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

```

### Ejercicio 5

Carga el dataset de accidentes de trfico en EEUU que est en la ruta "./data/car_crashes.csv". Quedate con los cinco estados con menos accidentes en "total".

```python
df_car_crashes = pd.read_csv("./data/car_crashes.csv", sep = "|") # cargamos dataset
df_car_crashes.head() # comprobamos si se ha cargado bien
```

```python
df_car_crashes = df_car_crashes.drop(columns=["Unnamed: 0"])  # eliminamos "Unnamed: 0" porque no nos aporta nada, solo ruido a nuestros datos.
```

```python
df_car_crashes.head() 
```

```python
accientes_por_estado = df_car_crashes.groupby("abbrev")["total"].sum().sort_values().reset_index() # agrupamos por Estado el nmero total de accidentes. En () ponemos el valor que agrupar. [] ponemos el campo que se tratar para hacer dichas agrupaciones. "sort_values" lo usamos para ordenar de menor a mayor.
top_5_estados_menos_accidentes = accientes_por_estado.head(5)                                       # Reset_index lo usamos para que el indice no se coma ninguna columna. De no poner esto la columna "abbrev" no se identificara.
top_5_estados_menos_accidentes
```

### Ejercicio 6

Completa el siguiente cdigo para obtener una figura como la representada en la imagen

```python
top_5_estados_menos_accidentes = pd.DataFrame(top_5_estados_menos_accidentes)
top_5_estados_menos_accidentes.columns
```

```python
#OJO df_top_min contiene el subdataframe resultado del ejercicio anterior
estados = {
    "DC": "District of Columbia",
    "MA": "Massachusetts",
    "MN": "Minnesota",
    "__": "Washington",
    "CT": "Connecticut"
}
for abrev,estado in estados.items():
    top_5_estados_menos_accidentes["abbrev"] = top_5_estados_menos_accidentes["abbrev"].str.replace(abrev,estado) # Utilizamos el bucle "for" para sustituir las abreviaturas por el nombre completo de los estados que hemos configurado en nuestro diccionario.
colores = ["green","gray","orange","blue","pink"] # establecemos la paleta de colores que queremos utilizar en nuestro grfico.
plt.figure(figsize = (15,5)) # configuramos nuestro lienzo , donde pintaremos nuestro grfico de barras.
plt.bar(top_5_estados_menos_accidentes["abbrev"], height = top_5_estados_menos_accidentes["total"], color = colores, alpha = 0.5) # plt.bar para crear un plot de barras. El primer componente sirve para decir qu es lo que definir cada una de las barras[Estados]. Height para establecer el alto de estas barras con los datos de los acciendetes(total de accidentes). Finalmente se establecen los colores y alpha el parmetro de transparencia.
plt.ylabel("Total Accidentes por Millones de Millas Recorridas") 
plt.ylim(0,30) # establecemos el lmite del eje Y 
plt.axhline(y = df_car_crashes.total.max()) # pintamos una linea con plt.axhline. Dicha linea viene dada por el valor mximo de accidentes .
plt.title("Ejercicio 6"); # Titulo de nuestro grfico de barras
```
